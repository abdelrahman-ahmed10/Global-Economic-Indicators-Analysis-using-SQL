```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
```

```python
database ='database.sqlite'
```

```python
conn = sqlite3.connect(database)
```

```python
query = """select *
                from sqlite_master
                where type='table';"""

tables = pd.read_sql(query, conn)
tables
```

```python
query = """select * from Indicators;"""
indicators = pd.read_sql(query, conn)
indicators.head()
```

```python
query = """SELECT COUNT(*) AS total_rows
FROM Indicators;"""
total_rows = pd.read_sql(query, conn)
total_rows
```

```python
query  = """SELECT COUNT(*) AS missing_values FROM Indicators
WHERE Value IS NULL;"""
null_value  = pd.read_sql(query, conn)
null_value
```

```python
query = """SELECT * FROM Indicators
WHERE Value IS NOT NULL
LIMIT 20;"""
not_null_values = pd.read_sql(query, conn)
not_null_values
```

```python
query = """SELECT *
FROM Indicators
WHERE CountryName IS NULL
   OR CountryCode IS NULL
   OR IndicatorName IS NULL
   OR IndicatorCode IS NULL
   OR Year IS NULL
   OR Value IS NULL;"""
missing_values = pd.read_sql(query, conn) 
missing_values
```

```python
query = """SELECT COUNT(DISTINCT IndicatorName) AS total_indicators
FROM Indicators;"""
total_indicators = pd.read_sql(query, conn)
total_indicators
```

```python
query = """SELECT DISTINCT IndicatorName
FROM Indicators
ORDER BY IndicatorName;"""
distinct_indicators = pd.read_sql(query, conn) 
distinct_indicators
```

```python
query = """SELECT DISTINCT IndicatorName
FROM Indicators
WHERE IndicatorName LIKE '%GDP%';"""
gdp_indicators = pd.read_sql(query, conn) 
gdp_indicators
```

```python
query = """SELECT Year, Value
FROM Indicators
WHERE CountryName = 'Brazil'
  AND IndicatorName = 'GDP per capita (current US$)'
ORDER BY Year DESC
LIMIT 10;"""
brazil_gdp = pd.read_sql(query, conn)
brazil_gdp
```

```python
qeury =  """ SELECT Year,CountryName,Value AS GDP_per_capita
FROM Indicators
WHERE CountryName IN ('Brazil', 'China')
  AND IndicatorName = 'GDP per capita (current US$)'
ORDER BY Year, CountryName;"""
brazil_china_gdp = pd.read_sql(qeury, conn)
brazil_china_gdp
```

```python
query = """SELECT Year,Value AS GDP_per_capita
FROM Indicators
WHERE CountryName = 'Brazil'
  AND IndicatorName = 'GDP per capita (current US$)'
  AND Year BETWEEN 1990 AND 1999
ORDER BY Year;"""
brazil_90s_gdp = pd.read_sql(query, conn)
brazil_90s_gdp
```

```python
query = """SELECT CountryName,IndicatorName, Value
FROM Indicators
WHERE CountryName IN ('Brazil', 'China', 'India')
  AND Year = 2014
  AND IndicatorName LIKE '%GDP%'
;"""
gdp_2014 = pd.read_sql(query, conn)
gdp_2014
```

```python
#Search for the Highest GDP Per Capita in 2014 o Find the country with the highest GDP per capita in 2014.
query = """SELECT CountryName, Value AS GDP_per_capita
FROM Indicators
WHERE IndicatorName = 'GDP per capita (current US$)'
  AND Year = 2014
ORDER BY Value DESC
;"""
highest_gdp_per_capita = pd.read_sql(query, conn)
highest_gdp_per_capita
```

```python
query = """SELECT CountryName,Value AS GDP_per_capita
FROM Indicators
WHERE IndicatorName = 'GDP per capita (current US$)'
  AND Year = 2014
ORDER BY Value DESC;"""
Reaffirm_highest_gdp_per_capita = pd.read_sql(query, conn)
Reaffirm_highest_gdp_per_capita
```

```python
query = """SELECT CountryName,
       AVG(Value) AS avg_GDP_per_capita
FROM Indicators
WHERE IndicatorName = 'GDP per capita (current US$)'
GROUP BY CountryName
ORDER BY avg_GDP_per_capita DESC;"""
average_gdp_per_capita = pd.read_sql(query, conn)
average_gdp_per_capita
```

```python
query = """SELECT COUNT(*) AS total_measures
FROM Indicators
WHERE IndicatorName = 'GDP per capita (current US$)'
  AND Year >= (SELECT MAX(Year) - 9 FROM Indicators);"""
total_measures = pd.read_sql(query, conn)
total_measures
```

```python
query = """SELECT CountryName,COUNT(*) AS measures_count
FROM Indicators
WHERE IndicatorName = 'GDP per capita (current US$)'
  AND CountryName = 'Angola'
GROUP BY CountryName;"""
angola_measures = pd.read_sql(query, conn)
angola_measures
```

```python
query = """SELECT 
    C.ShortName AS Country,
    I.IndicatorName,
    I.Year,
    I.Value
FROM Indicators I
JOIN Country C
    ON I.CountryCode = C.CountryCode
JOIN Series S
    ON I.IndicatorCode = S.SeriesCode
WHERE C.ShortName = 'Brazil'
  AND I.Year = 2014
  AND I.IndicatorName LIKE '%GDP%'
ORDER BY I.IndicatorName;"""
brazil_gdp_indicators_2014 = pd.read_sql(query, conn)
brazil_gdp_indicators_2014
```
